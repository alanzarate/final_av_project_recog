```python
cap  = cv2.VideoCapture('http://192.168.100.39:8080/video')
```

```python
def writeFrame(frame):
    dirPath = "./prod"
    frame_name = str(uuid.uuid4())
    frame_filename = os.path.join(dirPath, f'frame_{frame_name}.jpg')
    cv2.imwrite(frame_filename, frame)

limit = 10  ## write None if not limit needed
 
while True:
    ret, frame = cap.read()
    resized = cv2.resize(frame, (600,400))
    cv2.imshow("Frame", resized)
    
    key = cv2.waitKey(1)
    if key == ord("q"):
        break
    time.sleep(60)
    writeFrame(frame)
    
    if limit is not None:
        limit = limit - 1
    
    if limit < 0:
         break
    

cap.release()
cv2.destroyAllWindows()
```
